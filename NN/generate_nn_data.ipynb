```python
def generate_paths(num_layers, neurons_per_layer):
    train_paths = []
    test_paths = []
    
    # Iterate through even-indexed layers (0, 2, 4, 6, 8, ...)
    for start_layer in range(0, num_layers - 2, 2):
        for start_neuron in range(neurons_per_layer):
            start_index = start_neuron + start_layer * neurons_per_layer
            middle_index = start_neuron + (start_layer + 1) * neurons_per_layer
            end_index = start_neuron + (start_layer + 2) * neurons_per_layer

            test_path = (start_index, middle_index, end_index)
            test_paths.append(test_path)
            assert neurons_per_layer % 2 == 0
            if start_neuron < neurons_per_layer//2:
                middle_neurons = list(range(neurons_per_layer//2))
            else:
                middle_neurons = list(range(neurons_per_layer//2, neurons_per_layer))

            for middle_neuron in middle_neurons:
                middle_index = middle_neuron + (start_layer + 1) * neurons_per_layer
                
                for end_neuron in range(neurons_per_layer):
                    end_index = end_neuron + (start_layer + 2) * neurons_per_layer
                    
                    if middle_neuron != start_neuron or end_neuron != start_neuron:
                        train_path = (start_index, middle_index, end_index)
                        train_paths.append(train_path)

    return train_paths, test_paths

num_layers = 21
neurons_per_layer = 4

train_paths, test_paths = generate_paths(num_layers, neurons_per_layer)
```

```python
def expand_train_paths(original_train_paths):
    expanded_train_paths = []
    for path in original_train_paths:
        expanded_train_paths.append((path[0], path[1]))  # First two elements
        expanded_train_paths.append((path[1], path[2]))  # Last two elements
        expanded_train_paths.append(path)  # Full path
    return set(expanded_train_paths)

expanded_train_paths = expand_train_paths(train_paths)
```

```python
# distractory - pokud dvojička tak nesmí být žádný node ze stejné vrstvy (1,5) nesmí být node z vrstvy ze které je 1°
# u trojičky chceme dávat půl na půl samply, kde by neměl být node z prostřední vrstvy (1, 5, 8) - nesmí být node z vrstvy kde je 5
# ale hodit si mincí jestli tam dát node z vrstvy kde je 1 nebo ne, a ten musí být z druhé poloviny vrstvy (2, 3)
# 0 1
# 2 3
# 2 6 8
# 0 1
```

```python
import random
def generate_to_exclude_dict(num_layers, neurons_per_layer):
    exclusion_dic = {}
    total_neurons = num_layers * neurons_per_layer

    for neuron in range(total_neurons):
        layer_start = (neuron // neurons_per_layer) * neurons_per_layer
        layer_end = layer_start + neurons_per_layer
        exclusion_dic[neuron] = list(range(layer_start, layer_end))

    return exclusion_dic

exclusion_dic = generate_to_exclude_dict(num_layers, neurons_per_layer)

def get_distractors(train_paths, exclusion_dic, N, num_of_combinations):
    train_data = []
    all_nodes = set(range(len(exclusion_dic)))
    for path in train_paths:
        for _ in range(0,num_of_combinations):
            if len(path) == 2:
                first_node, second_node = path
                to_exclude = set(exclusion_dic[first_node])

                available_nodes = list(all_nodes - to_exclude)
                distractors = random.sample(available_nodes, N+1)

                input_list = [first_node] + distractors
                random.shuffle(input_list)
                input_list.append(second_node)
                orig_idx = input_list.index(first_node)

                sample = {"input": input_list, "orig_idx": orig_idx, "target_idx": orig_idx + 84}
                train_data.append(sample)
            elif len(path) == 3:
                first_node, middle_node, last_node = path
                
                to_exclude = set(exclusion_dic[middle_node] + exclusion_dic[first_node])
                
                available_nodes = list(all_nodes - to_exclude)

                if random.choice([True, False]):
                    layer_start = (first_node // neurons_per_layer) * neurons_per_layer
                    layer_mid = layer_start + neurons_per_layer // 2
                    if first_node < layer_mid:  # First node is in the first half
                        sample_range = range(layer_mid, layer_start + neurons_per_layer)
                    else:  # First node is in the second half
                        sample_range = range(layer_start, layer_mid)
                    
                    sampled_node = random.choice(sample_range)
                    distractors = random.sample(available_nodes, N-1)
                    distractors += [sampled_node]
                else:
                    distractors = random.sample(available_nodes, N)

                input_list = [first_node] + [middle_node] + distractors
                
                random.shuffle(input_list)
                input_list.append(last_node)
                orig_idx = input_list.index(first_node)
                sample = {"input": input_list, "orig_idx": orig_idx, "target_idx": orig_idx + 84}
                train_data.append(sample)
            
    return train_data
```

```python
def get_test_distractors(test_paths, exclusion_dic, N):
    test_data = []
    all_nodes = set(range(len(exclusion_dic)))
    for path in test_paths:
        first_node, middle_node, last_node = path
        
        to_exclude = set(exclusion_dic[middle_node] + exclusion_dic[first_node])
        
        available_nodes = list(all_nodes - to_exclude)

        if random.choice([True, False]):
            layer_start = (first_node // neurons_per_layer) * neurons_per_layer
            layer_mid = layer_start + neurons_per_layer // 2
            if first_node < layer_mid:  # First node is in the first half
                sample_range = range(layer_mid, layer_start + neurons_per_layer)
            else:  # First node is in the second half
                sample_range = range(layer_start, layer_mid)
            
            sampled_node = random.choice(sample_range)
            distractors = random.sample(available_nodes, N-1)
            distractors += [sampled_node]
        else:
            distractors = random.sample(available_nodes, N)

        input_list = [first_node] + [middle_node] + distractors
        
        random.shuffle(input_list)
        input_list.append(last_node)
        orig_idx = input_list.index(first_node)

        sample = {"input": input_list, "target_idx": orig_idx + 84, "orig_idx": orig_idx}
        test_data.append(sample)

    return test_data
```

```python
train_data = get_distractors(train_paths, exclusion_dic, 17, 929)
test_data = get_test_distractors(test_paths, exclusion_dic, 17)
```
